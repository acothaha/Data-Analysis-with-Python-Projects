```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
df = pd.read_csv('epa-sea-level.csv')
df.head()
```

```python
fig, ax = plt.subplots(figsize=(14,8))

df.plot(kind='scatter', x='Year', y='CSIRO Adjusted Sea Level', ax=ax);
```

```python
fig, ax = plt.subplots(figsize=(14,8))
year_add1 = range(1880,2051)
year_add2 = range(2000,2051)


slope1, intercept1, r_value1, p_value1, std_err1 = linregress(df['Year'], df['CSIRO Adjusted Sea Level'])
slope2, intercept2, r_value2, p_value2, std_err2 = linregress(df['Year'][120:], df['CSIRO Adjusted Sea Level'][120:])
df.plot(kind='scatter', x='Year', y='CSIRO Adjusted Sea Level', ax=ax);
ax.plot(year_add1, intercept1 + slope1*year_add1, 'black', label='fitted line (from 1880)')
ax.plot(year_add2, intercept2 + slope2*year_add2, 'r', label='fitted line (from 2000)')
ax.set_ylabel('Sea Level (inches)')
ax.set_title('Rise in Sea Level')
ax.legend()
plt.show()
```
